## Problem 3
Determine lowest benzene composition in the bottoms given certain conditions.

$15$ trays at $20$% efficiency is equivalent to $3$ equilibrium stages.

The minimum composition in the bottoms is achieved at total reflux where the operating line is $y=x$.

## Problem 3.1
Total condensers and total reboilers do not count as equilibrium stages so the column still only has 3 equilibrium stages.

```python
import pandas as pd
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('Txy.csv')                       #data from csv
temp = file['T'].values
x = file['x'].values
y = file['y'].values

xInterp = interp1d(y,x)                             #interpolated data
yInterp = interp1d(x,y)

z = .5                                              #feed comp molfrac
F = 45                                              #feed kg/hr
xD = .8                                            #molfrac distil
```

```python
plt.figure(figsize = (9,7))
plt.plot(x,y,label='EQ curve')
plt.plot(x,x,label='Operating line')
plt.vlines(xD,0,1,'k','--')

x1 = xD
y1 = xD
x2 = xInterp(y1)
y2 = x2
x3 = xInterp(y2)
y3 = x3
x4 = xInterp(y3)
y4 = x4

plt.hlines(y1,x1,x2)
plt.vlines(x2,y1,y2)
plt.hlines(y2,x2,x3)
plt.vlines(x3,y2,y3)
plt.hlines(y3,x3,x4)
plt.vlines(x4,y3,y4)


plt.grid()
plt.title('Total Reflux')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
;
print(x4)
```

The minimum composition in the bottoms with a total reboiler and total condenser is $x_B=0.225$.

## Problem 3.2
With a partial condenser and total reboiler, there is one added equilibrium stage to the column so $N=4$.

```python
plt.figure(figsize = (9,7))
plt.plot(x,y,label='EQ curve')
plt.plot(x,x,label='Operating line')
plt.vlines(xD,0,1,'k','--')

x1 = xD
y1 = xD
x2 = xInterp(y1)
y2 = x2
x3 = xInterp(y2)
y3 = x3
x4 = xInterp(y3)
y4 = x4
x5 = xInterp(y4)
y5 = x5

plt.hlines(y1,x1,x2)
plt.vlines(x2,y1,y2)
plt.hlines(y2,x2,x3)
plt.vlines(x3,y2,y3)
plt.hlines(y3,x3,x4)
plt.vlines(x4,y3,y4)
plt.hlines(y4,x4,x5)
plt.vlines(x5,y4,y5)

plt.grid()
plt.title('Total Reflux')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
;
print(x5)
```

The minimum composition with a partial condenser and total reboiler is $x_B=0.112$.
